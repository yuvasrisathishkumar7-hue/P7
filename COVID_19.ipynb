**OBJECTIVE 9 – DASHBOARD DEVELOPMENT**

**20 – Dashboard Layout**

```python
app = JupyterDash(__name__)

app.layout = html.Div([

    html.H1("COVID-19 Dashboard", style={"textAlign": "center"}),

    dcc.Dropdown(
        id="country",
        options=[{"label": i, "value": i} for i in df["Country"]],
        value=df["Country"].iloc[0]
    ),

    # Dynamic Country Graph
    dcc.Graph(id="graph"),

    # Bar Chart – Top affected countries
    dcc.Graph(
        figure=px.bar(
            top10,
            x="Country",
            y="Confirmed",
            title="Top 10 Countries"
        )
    ),

    # Line Chart – Cases trend
    dcc.Graph(
        figure=px.line(
            top10,
            x="Country",
            y="Confirmed",
            markers=True,
            title="Cases Trend (Line Chart)"
        )
    ),

    # Scatter Plot – Relationship
    dcc.Graph(
        figure=px.scatter(
            df,
            x="Confirmed",
            y="Deaths",
            size="Active",
            color="Recovered",
            hover_name="Country",
            title="Confirmed vs Deaths"
        )
    ),

    # Heatmap – Correlation
    dcc.Graph(
        figure=px.imshow(
            df.drop("Country", axis=1).corr(),
            text_auto=True,
            title="Correlation Matrix"
        )
    ),

    # Pie Chart – Case Distribution
    dcc.Graph(
        figure=px.pie(
            pie_data,
            values="Confirmed",
            names="Country",
            title="Case Distribution (Top 5 + Others)"
        )
    )

])
```

```python
@app.callback(
    Output("graph", "figure"),
    Input("country", "value")
)
def update_graph(selected_country):

    data = df[df["Country"] == selected_country]

    fig = px.bar(
        x=["Confirmed","Deaths","Recovered","Active"],
        y=[
            data["Confirmed"].values[0],
            data["Deaths"].values[0],
            data["Recovered"].values[0],
            data["Active"].values[0]
        ],
        title=f"COVID Data - {selected_country}"
    )

    return fig
```

```python
app.run()
```
